```python
import numpy as np
import matplotlib.pyplot as plt
```

Defining the Given Function

```python
def linear_func(w,x,b):
  return w*x + b

def square_func(x):
  return x**2

def sigmoid_func(x):
  return 1/(1+np.exp(-x))

def tanh_func(x):
  return ((np.exp(x) - np.exp(-x))/(np.exp(x) - np.exp(-x)))

def composite_func1(w,x,b):
  u = linear_func(w,x,b)
  return sigmoid_func(u)

def composite_func2(w,x,b):
  u = linear_func(w,x,b)
  return tanh_func(u)

def composite_func3(w1,w2,w3,w4,x,b,b1,b2):
  u1 = linear_func(w1,x,b1)
  y1 = sigmoid_func(u1)
  u2 = linear_func(w2,x,b2)
  y2 = sigmoid_func(u2)

  v = w3 * y1 + w4 * y2 + b
  return  sigmoid_func(v)
```

Taking x values for given range

```python
x = np.arange(-10,10,0.1)
```

Taking weights and biases

```python
w = 2
w1 = .5
w2 = -0.5
w3 = 2
w4 = 1
b =3
b1 = -1
b2 = 1
```

Calculating Y for corresponding functions

```python
y_linear = linear_func(w,x,b)
y_square = square_func(x)
y_sig = sigmoid_func(x)
y_tanh = tanh_func(x)
y_comp1 = composite_func1(w,x,b)
y_comp2 = composite_func2(w,x,b)
y_comp3 = composite_func3(w1,w2,w3,w4,x,b,b1,b2)
```

Ploting Graphs for every functions (x,y_func)

```python
plt.figure(figsize=(20,15))

plt.subplot(421)
plt.plot(x,y_linear, label="Linear Function")
plt.title("Linear Function")
plt.legend()

plt.subplot(422)
plt.plot(x,y_square, label="Square Function")
plt.title("Square Function")
plt.legend()

plt.subplot(423)
plt.plot(x,y_sig, label="Sigmoid Function")
plt.title("Sigmoid Function")
plt.legend()

plt.subplot(424)
plt.plot(x,y_tanh, label="Tanh Function")
plt.title("Tanh Function")
plt.legend()

plt.subplot(425)
plt.plot(x,y_comp1, label="Composite - 1")
plt.title("Composite Function 1")
plt.legend()


plt.subplot(426)
plt.plot(x,y_comp2, label="Composite - 2")
plt.title("Composite Function 2")
plt.legend()


plt.subplot(427)
plt.plot(x,y_comp3, label="Composite - 3")
plt.title("Composite Function 3")
plt.legend()

plt.tight_layout()
```
