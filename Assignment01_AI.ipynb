```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.random.uniform(-100,100)
b = np.random.uniform(-100,100)
x = np.random.uniform(-10,10,50)

# print(a,b,x)
```

```python
def linear(a,b,x):
  return a * x + b

def exponential(a,b,x):
  return a* np.exp(x) + b

def quadratic(a,b,x):
  return a * x** 2 +b
```

```python
linear_y = []
exp_y = []
quad_y = []

linear_y.append(linear(a,b,x))
linear_y = np.squeeze(linear_y)
exp_y.append(exponential(a,b,x))
exp_y = np.squeeze(exp_y)
quad_y.append(quadratic(a,b,x))
quad_y = np.squeeze(quad_y)
```

```python
plt.scatter(x,linear_y, color='grey')
plt.title("Linear: y = {}x + {}".format(round(a,2), round(b,2)))
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.scatter(x,exp_y, color='red')
plt.title("Exponential: y = {}e^x + {}".format(round(a,2), round(b,2)))
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.scatter(x,quad_y, color='red')
plt.title("Quadratic: y = {}x^2 + {}".format(round(a,2), round(b,2)))
plt.xlabel('x')
plt.ylabel('y')
```
